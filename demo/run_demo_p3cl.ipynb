# Detecting cell subpopulation and club on cell line mixture p3cl dataset

This example shows the usage of `pyseat.SEAT` to detect cell subpopulation and club on cell line mixture p3cl dataset.

```python
%reload_ext autoreload
%autoreload 2

import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn import datasets 
import umap

from pyseat.SEAT import SEAT

# load data
X = pd.read_csv('p3cl_count_umap.csv', index_col=0)
meta_df = pd.read_csv('p3cl_metadata.csv')
meta_df.index = X.index

# run seat
seat = SEAT(affinity="gaussian_kernel",
            sparsification="knn_neighbors_from_X",
            n_neighbors=30,
            objective="SE",
            strategy="bottom_up")

seat_clusters = seat.fit_predict(X)

# seat output
print('--------------------------')
print('The best k is:', seat.optimal_k)
print('SEAT optimal clustering result: \n', seat.labels_)
print('Candicate k for tuning: \n', seat.ks)
print('SE score for tuning k: \n', seat.se_scores)
print('SEAT clustering result for different k: \n', seat.ks_clusters)
print('SEAT club labels: \n', seat.clubs)
print('SEAT orders: \n', seat.order)

# plot the structure entropy hierarchy and clustering result
label = seat.labels_
cluster_df = seat.ks_clusters
cluster_df.index = X.index
cluster_df['Club'] = seat.clubs
cluster_df['Subpopulation'] = seat.labels_
cluster_df['Ground Truth'] = meta_df['cell_line']
for col in cluster_df.columns:
    group = list(set(cluster_df[col]))
    label_colors = dict(zip(group, sns.color_palette('Spectral', len(group)).as_hex()))
    cluster_df[col] = cluster_df[col].map(label_colors)

sns.clustermap(X, cmap='YlGnBu', col_cluster=False, row_cluster=False)
plt.title("Original data")
plt.show()
sns.clustermap(X,
                   row_linkage=seat.Z_,
                   col_cluster=False,
                   row_colors=cluster_df,
                   cmap='YlGnBu')
plt.title("After SEAT clustering and ordering")
plt.show()
```

# Hierachical embedding on cell line mixture p3cl dataset

This example shows the usage of `pyseat.HierachicalEmbedding` to visualize the nested structure of cells on cell line mixture p3cl dataset.


```python
import numpy as np
import pandas as pd
import umap
from matplotlib.colors import ListedColormap
from pyseat.HierachicalEmbedding import HierachicalEmbedding

y = pd.DataFrame({
    'Global': [1]*len(seat.labels_),
    'Subpopulation': seat.labels_,
    'Club': seat.clubs
})
y.index = X.index

HE = HierachicalEmbedding(device='cpu', n_epochs=1000, init='spectral', random_state=0,
                          min_dist=0.1,
                          n_components=2)  

embed = HE.fit_transform(seat.aff_m, y, thetas=[1, 1, 1])
print('SEAT hierachical embedding: ', embed.shape)
HE.viz_fit()

color_set = {x:i for i, x in enumerate(set(meta_df['cell_line']))}
y['Ground Truth'] = [color_set[x] for x in meta_df['cell_line']]
plt.title('HE, colored by Ground Truth')
plt.scatter(embed[:, 0], embed[:, 1], 
                s=1,
                c=y['Ground Truth'], 
                cmap=ListedColormap(sns.color_palette('Spectral', len(set(y['Ground Truth']))))
               )
plt.show()

# plot umap embedding
umap_embed = umap.UMAP(random_state=0).fit_transform(seat.aff_m)
plt.figure(figsize=(20, 5))
fig_count = 1

for i in range(y.shape[1]):
    plt.subplot(1, 4, fig_count)
    plt.title('UMAP, colored by {}'.format(y.columns[i]))
    plt.scatter(umap_embed[:, 0], umap_embed[:, 1], 
                s=1,
                c=y[y.columns[i]], 
                cmap=ListedColormap(sns.color_palette('Spectral', len(set(y[y.columns[i]]))))
               )
    fig_count += 1
plt.show()
```
